```python
#Tipo de triángulo

lado1 = int(input("Ingrese el primer lado: "))
lado2 = int(input("Ingrese el segundo lado: "))
lado3 = int(input("Ingrese el tercer lado: "))

if lado1 == lado2 and lado2 == lado3:
  print("El triángulo es equilátero")
elif lado1 == lado2 or lado1 == lado3 or lado2 == lado3:
  print("El triángulo es isóceles")
else:
  print("El triángulo es escaleno")
```

```python
# Calculadora simple

num1 = float(input("Ingrese el primer número: "))
num2 = float(input("Ingrese el segundo número: "))
operacion = input("Ingrese la operación (+, -, *, /): ")

if operacion == "+":
    resultado = num1 + num2

elif operacion == "-":
    resultado = num1 - num2

elif operacion == "*":
    resultado = num1 * num2

elif operacion == "/":
    if num2 != 0:
        resultado = num1 / num2
    else:
        print("No se puede dividir entre 0")
        resultado = None
else:
    print("Operación no válida")
    resultado = None

if resultado != None:
    print("Resultado:", resultado)
```

```python
# Convertir nota numérica a letra

nota = int(input("Ingrese la nota (0-100): "))

if nota >= 90 and nota <= 100:
    print("Calificación: A")

elif nota >= 80:
    print("Calificación: B")

elif nota >= 70:
    print("Calificación: C")

elif nota >= 60:
    print("Calificación: D")

elif nota >= 0:
    print("Calificación: F")

else:
    print("Nota inválida")
```

```python
# Validar triángulo y determinar tipo

a = float(input("Ingrese el primer lado: "))
b = float(input("Ingrese el segundo lado: "))
c = float(input("Ingrese el tercer lado: "))

# Verificar si puede formar triángulo
if a + b > c and a + c > b and b + c > a:

    if a == b and b == c:
        print("Sí forman un triángulo equilátero")

    elif a == b or a == c or b == c:
        print("Sí forman un triángulo isósceles")

    else:
        print("Sí forman un triángulo escaleno")

else:
    print("No pueden formar un triángulo")
```
